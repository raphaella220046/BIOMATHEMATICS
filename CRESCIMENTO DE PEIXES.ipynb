##### A equação diferencial do Modelo de Von Bertalanffy é:

$$\large\frac{dp}{dt}+\beta p= a p^{\frac{2}{3}}$$

Essa equação pode ser aplicada ao crescimento de peixes, estabelecendo uma relação proporcional entre seu peso e sua superfície corporal ao longo do tempo.

Etapas para aplicar as simplificações de Bernoulli

- Dividir ambos os lados da equação por $\large p^{\frac{2}{3}}$

$$\large \frac{dp}{dt}p^{-\frac{2}{3}}+\beta p \cdot p^{-\frac{2}{3}} = a \cdot p^{\frac{2}{3}} \cdot p^{-\frac{2}{3}}$$

logo,

$$\large \frac{dp}{dt}p^{-\frac{2}{3}}+\beta \cdot p^{\frac{1}{3}} = a$$

- Fazer a mudança de variável $\large z=p\frac{1}{3}$

$$\large \frac{dz}{dt}=\frac{1}{3}p^{\frac{1}{3}-1} \frac{dp}{dt} \implies \large \frac{dz}{dt}=\frac{1}{3}p^{-\frac{2}{3}} \frac{dp}{dt} \implies 3 \frac{dz}{dt}=p^{-\frac{2}{3}}\frac{dp}{dt}$$

- Substituir na equação anterior

$$\large \frac{dp}{dt}p^{-\frac{2}{3}}+\beta \cdot p^{\frac{1}{3}} = a \implies 3 \frac{dz}{dt} + \beta z = \boxed{a \implies \frac{dz}{dt}=\frac{a-\beta z}{3}}$$

Etapas para resolver a EDO pelo método de separação de variáveis

- Separar $dz$ e $dt$

$$\large \frac{3}{a-\beta z}dz=dt$$

- Isolar a constante

$$\large 3\int \frac{1}{a-\beta z}dz=\int dt$$

- Aplicar a substituição pelo método udu, em que $\large u=a-\beta z$ e $\large \frac{du}{dz}=-\beta \implies dz=-\frac{1}{\beta}du$

$$\large -\frac{3}{\beta}\int \frac{1}{u}du=\int t$$

- Fazer a integração

$$\large -\frac{3}{\beta}ln(a-Bz)=t+c$$

- Colocar todos os termos da equação na base Euler e aplicar as propriedades

$$\large e^{-\frac{3}{\beta}}e^{ln(a-\beta z)}=e^{t+c}$$

- Para $\large e^{t+c}=e^t\cdot e^c$ e $\large e^c=A$, então

$$\large e^{-\frac{3}{\beta}}e^{ln(a-\beta z)}=e^t A$$

Em que $e^{ln(x)}=x$, visto que a base do expoente é igual a base do logaritmo, assim

$$\large a-\beta z=e^{-\frac{\beta t}{3}}A$$

- Isolar z para obter a solução

$$\large \boxed{z=\frac{a-A}{\beta}e^{-\frac{\beta t}{3}}}$$

Os pontos de equilíbrio ocorrem quando a derivada é nula, ou seja, $\frac{dp}{dt}=0$, logo

$$\large \frac{dp}{dt}+\beta p=ap^{\frac{2}{3}}$$

$$\large \frac{dp}{dt}=ap^{\frac{2}{3}}-\beta p=0$$

- Colocar p em evidência

$$\large p(ap^{-\frac{1}{3}}-\beta)=0$$

- Nesse sentido, para o campo vetorial ser nulo, ou $p=0$ que é uma solução trivial ou $p$ é igual a

$$\large ap^{-\frac{1}{3}}=\beta \implies \sqrt[3]{\frac{1}{p}}=\frac{\beta}{a} \implies \boxed{p=\left(\frac{a}{\beta}\right)^3}$$

- Para descobrir se o ponto de equilíbrio é assintoticamente estável ou instável, basta fazer a derivação

$$\large \frac{dp}{dt}=ap^{\frac{2}{3}}-\beta p \implies \frac{2}{3}ap^{-\frac{1}{3}}-\beta$$

- E em sequência, a substituição de $\large p=\left(\frac{a}{\beta}\right)^3$

$$\large \frac{2}{3}ap^{-\frac{1}{3}}-\beta \implies \frac{2}{3} a \left( \left( \frac{a}{\beta} \right)^3 \right)^{-\frac{1}{3}}-\beta \implies \frac{2}{3} a \left( \frac{\beta}{a} \right)-\beta=-\frac{\beta}{3}<0$$

Portanto, o ponto $\large p=\left(\frac{a}{\beta}\right)^3$ é assintoticamente estável.

```python
# Importações
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Parâmetros do modelo de Von Bertalanffy
a = 0.1 # constante de anabolistmo
b = 0.025 # constante de catabolismo

# Função de Von Bertalanffy
def modelo_Bertalanffy(p, t):
    dpdt = (a*p**(2/3))-(b*p)
    return dpdt

# Ponto de estabilidade
p_estabilidade = (a / b) ** 3

# Valores de p (peso do peixe) e t (tempo)
p = np.linspace(0, 2 * p_estabilidade, 20)
dp_dt = np.linspace(-1, 1, 20)

# Definição do campo vetorial
P, DP_DT = np.meshgrid(p, dp_dt)
dP_dt = modelo_Bertalanffy(P, DP_DT)

# Representação do campo vetorial
plt.figure(figsize = (8, 6))
plt.quiver(P, DP_DT, np.ones_like(dP_dt), dP_dt, scale=20, color='blue', label='dp/dt = a * p^(2/3) - b * p')
plt.plot(p_estabilidade, 0, 'ro', markersize=8, label=f'Ponto de Estabilidade (p = {p_estabilidade:.1f})')
plt.xlabel('p')
plt.ylabel('dp/dt')
plt.title('Campo Vetorial do Modelo de Von Bertalanffy')
plt.legend()
plt.grid()
plt.show()
```

O modelo de Von Bertalanffy descreve como o tamanho do organismo muda em relação ao tempo com base nas taxas de crescimento e mortalidade, entendendo que o crescimento de um organismo não é constante ao longo de sua vida, sugerindo que o crescimento é rápido quando o organismo é jovem e saudável, mas diminui à medida que este organismo envelhece ou enfrenta condições adversas.

O campo vetorial para o modelo de Von Bertalanffy apresenta um ponto de inflexão, demonstrando essa não linearidade do peso do peixe no decorrer do tempo. É valido observar que a derivada $\large \frac{dp}{dt}=ap^{\frac{2}{3}}-\beta p$ tem como segunda derivada $\large \frac{d^2p}{dt}=\frac{2}{3}ap^{-\frac{1}{3}}-\beta$. Igualando a zero, tem-se $\large \frac{2}{3}ap^{-\frac{1}{3}}-\beta=0 \implies \sqrt[3]{\frac{1}{p}}=\frac{3\beta}{2a} \implies p=\left( \frac{2a}{3\beta} \right)^3$, em que o peso do peixe  representa um ponto de inflexão com concavidade para cima, desde que $\frac{d^2p}{dt}>0$, assumindo que o valor de $a$ seja maior que $\beta$ antes da derivada atingir seu ponto de máximo quando se iguala a zero.
