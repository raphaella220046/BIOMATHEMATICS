O modelo de Lotka-Volterra consiste em equações diferenciais não lineares que descrevem a dinâmica de populações de presas e predadores em um ecossistema. Ele foi proposto por Alfred Lotka e Vito Volterra na década de 1920.

O modelo considera duas espécies interagindo no mesmo ecossistema:

1. **População de Presas ($x$):** Representa o número de indivíduos da espécie presa.

2. **População de Predadores ($y$):** Representa o número de indivíduos da espécie predadora.

As equações do modelo de Lotka-Volterra são descritas por:

$$
\begin{align*}
    \begin{cases}
        \frac{dx}{dt} = a x - \alpha xy \\
        \frac{dy}{dt} = -b y + \beta xy \\
    \end{cases}
\end{align*}
$$

Em que:

- $\frac{dx}{dt}$ e $\frac{dy}{dt}$ são as taxas de variação das populações de presas e predadores em relação ao tempo, respectivamente.

- $a$ é a taxa de crescimento natural das presas na ausência de predadores.

- $\alpha$ é a taxa de predação, ou seja, a eficiência com que os predadores capturam e consomem as presas.

- $b$ é a taxa de mortalidade natural dos predadores na ausência de presas.

- $\beta$ é a taxa de conversão, ou seja, a eficiência com que os predadores aumentam sua população ao consumir presas.

O modelo assume algumas simplificações, como a ausência de fatores ambientais externos, migração, e capacidade de carga do ambiente.

A dinâmica do modelo mostra um ciclo de oscilação entre as populações de presas e predadores, onde o aumento da população de presas leva a um aumento subsequente na população de predadores, que, por sua vez, leva a uma redução nas presas, criando um ciclo contínuo. O modelo sugere que as flutuações nas populações de presas e predadores estão interligadas e influenciam umas às outras ao longo do tempo.

#### A separação de variáveis

As equações diferenciais podem ser estudadas em termos de seu comportamento no que se refere a estabilidade, aplicando a separação de variáveis

$$\frac{dx}{dt} \frac{dt}{dy} = \frac{ax-\alpha xy}{-by+ \beta xy} \implies \frac{dx}{dy} = \frac{x(a- \alpha y)}{y(-b+\beta x)} \implies \frac{(-b+\beta x) }{x}dx = \frac{(a-\alpha y)}{y}dy$$

Integrando em ambos os lados da equação

$$\int \frac{(-b+\beta x)}{x}dx = \int \frac{(a-\alpha y)}{y}dy$$

$$\int \frac{-b}{x}dx + \int \beta dx = \int \frac{a}{y} dy - \int \alpha dy$$

$$-b ln(x) + \beta x = a ln(y) - \alpha y + c$$
$$ b ln(x) - \beta x + a ln(y) - \alpha y = c $$

Na qual $c$ é a constante de integração.

Aplicando a propriedade dos logaritmos e separando os termos contendo $x$ e $y$, tem-se

$$ln y^a - \alpha y = ln x^{-b} + \beta x + ln C,$$

em que $ln C = c.$

Aplicando a exponencial em ambos os lados, é obtido

$$y^a e^{-\alpha y} = C e^{\beta x} x^{-b}.$$

Desta forma, a constante C pode ser determinada de acordo com as condições iniciais ($x_0, y_0$) em

$$C = \frac{y_0^a e^{-\alpha y_0}}{e^{\beta x_0} x_0^{-b}}$$

Rearranjando os termos e passando o termo do denominador para o numerador ao multiplicar o expoente por $-1$, chega-se a equação

$$C = x_0^b y_0^a e^{-\beta x_0} e^{-\alpha y_0}$$

#### Os pontos de equilíbrio

É possível obter os pontos de equilíbrio do modelo ao igualar as derivadas a zero

$$
\begin{align*}
    \begin{cases}
        \frac{dx}{dt} = a x - \alpha xy = 0 \\
        \frac{dy}{dt} = -b y + \beta xy = 0 \\
    \end{cases}
\end{align*}
$$

As condições que satisfazem essa igualdade são

$$x = 0$$ e 
$$y = 0.$$

Se $x > 0$ e $y > 0$. Outra possibilidade é fazer o termo entre parêntese igual a zero

$$a - \alpha y = 0 \implies y = \frac{a}{\alpha},$$
$$-b + \beta x = 0 \implies x = \frac{b}{\beta}.$$

Logo, os dois pontos de equilíbrio são dados por $P_1 = (0,0)$ e $P_2 = \left( \frac{b}{\beta}, \frac{a}{\alpha} \right)$.

#### Tratando o modelo por sistema linear

As equações diferenciais no modelo são compostas por duas partes

$$
\begin{align*}
    \begin{cases}
        \frac{dx}{dt} = (a x) - (\alpha xy) = (\text{linear}) - (\text{não linear}), \\
        \frac{dy}{dt} = -b y + \beta xy = -(\text{linear}) + (\text{não linear}). \\
    \end{cases}
\end{align*}
$$

Dessa forma, é possível analisar o sistema linear

$$
\begin{align*}
    \begin{cases}
        \frac{dx}{dt}=ax=0 \\
        \frac{dy}{dt}=-bx=0 \\
    \end{cases}
\end{align*}
$$

na forma matricial.

#### Estabilidade dos pontos de equilíbrio

Considerando o ponto $P_1 = (0,0)$

$$
\begin{align*}
A\cdot x & = b \\
 \begin{bmatrix}
     a & 0 \\
     0 & -b
 \end{bmatrix}
 \cdot
 \begin{bmatrix}
 x \\
 y
 \end{bmatrix}
 &=
 \begin{bmatrix}
 x' \\
 y'
 \end{bmatrix}
\end{align*}
$$

Resolvendo o determinante $det|A- \lambda I| = 0$

$$
\begin{align*}
    \begin{bmatrix}
        a- \lambda & 0 \\
        0 & -b - \lambda
    \end{bmatrix}
    &=
    \begin{bmatrix}
     0 \\
     0
    \end{bmatrix}
\end{align*}
$$

$$(a-\lambda)(-b-\lambda)=0$$
$$\lambda^2 - a\lambda + b\lambda - ab = 0$$

Substituindo os parâmetros $a=0,08$ e $b=-0,02$.

$$\lambda^2 - 0,08\lambda -0,02\lambda + 0,0016 = 0$$

$$ \lambda^2 - 0,1 \lambda + 0,0016 = 0$$

$$\lambda = \frac{0,1 \pm \sqrt{0,01 - 0,0064}}{2}$$

$$\lambda=\frac{0,1 \pm 0,06}{2}$$

Logo, os autovalores podem ser $\lambda_1 = 0,08$ e $\lambda_2=0,02$, e os respectivos autovetores são

$\nu_1 = \begin{bmatrix}
    1 \\
    0 \\
\end{bmatrix}$ e 

$\nu_2 = \begin{bmatrix}
    0 \\
    1 \\
\end{bmatrix}.$

E a solução na forma geral é dada por

$$
\begin{bmatrix}
    x \\
    y \\
\end{bmatrix}
=c_1
\begin{bmatrix}
    1 \\
    0 \\
\end{bmatrix}
e^{at} + c
\begin{bmatrix}
    0 \\
    1 \\
\end{bmatrix}
e^{-bt}.
$$

Aplicando a segunda derivada nas equações diferenciais também é possível obter a informação sobre a estabilidade nesse ponto

$$
\begin{align*}
    \begin{cases}
        \frac{d^2x}{dt^2}=a-\alpha y = a - \alpha \cdot 0 = a = 0,08 \\
        \frac{d^2y}{dt^2} = -b + \beta x = -b + \beta \cdot 0 = -b = 0,02 \\
    \end{cases}
\end{align*}
$$

Portanto, o ponto $P_1 = (0,0)$ é instável, visto que a segunda derivada, assim como os autovalores da matriz são positivos.

#### Analisando a Matriz Jacobiana

A Matriz Jacobiana é dada por

$$
\begin{bmatrix}
     \frac{df}{dx} & \frac{df}{dy} \\
     \frac{dg}{dx} & \frac{dg}{dy}
 \end{bmatrix}
$$

Assumindo que

$$
\begin{align*}
    \begin{cases}
        \frac{dx}{dt} = f(x,y) \\
        \frac{dy}{dt} = g(x,y) \\
    \end{cases}
\end{align*}
$$

A Matriz Jacobiana pode ser calculada como

$$
J(x,y) = \begin{bmatrix}
     a - \alpha y & - \alpha x \\
     \beta y & -b + \beta x
 \end{bmatrix}
$$

No ponto $P_1 = (0,0)$ a solução da matriz é do tipo

$$
\begin{align*}
J(0,0) = \begin{bmatrix}
     a - \alpha y & - \alpha x \\
     \beta y & -b + \beta x
\end{bmatrix}
&=
\begin{bmatrix}
     a & 0 \\
     0 & -b
\end{bmatrix}
\end{align*}
$$

Em que os autovalores se caracterizam por ter um comportamento instável com ponto de sela, como mostra a figura 1.

Figura 1: Plano de fase apresentando ponto de sela. Fonte: HENAREJOS, 2023 [1]

Calculando J no ponto $P_2 = \left( \frac{b}{\beta}, \frac{a}{\alpha} \right)$ é obtido o sistema linear

$$
\frac{d}{dt}
\begin{bmatrix}
      x \\
      y
 \end{bmatrix}
 =
\begin{bmatrix}
     0 & - \alpha \frac{b}{\beta} \\
     \beta \frac{a}{\alpha} & 0
 \end{bmatrix}
\cdot
\begin{bmatrix}
      x \\
      y
 \end{bmatrix}
$$

Em que os autovalores se caracterizam por ter um comportamento estável com um ponto de centro, como mostra a figura 2.

Figura 2: Plano de fase apresentando ponto de centro. Fonte: HENAREJOS, 2023 [1]

Fazendo a mudança de variável $u=x-\frac{b}{\beta}$ e $w=y-\frac{a}{\alpha}$ tem-se

$$
\frac{d}{dt}
\begin{bmatrix}
      u \\
      w
 \end{bmatrix}
 =
\begin{bmatrix}
     0 & - \alpha \frac{b}{\beta} \\
     \beta \frac{a}{\alpha} & 0
 \end{bmatrix}
\cdot
\begin{bmatrix}
      u \\
      w
 \end{bmatrix}
$$

Substituindo o ponto de equilíbrio $P_2 = (1000, 80)$ na matriz jacobiana e considerando $u=x-1000$ e $w=y-80$ tem-se

$$
\frac{d}{dt}
\begin{bmatrix}
      u \\
      w
 \end{bmatrix}
 =
\begin{bmatrix}
     0 & - 1000 \alpha \\
     80 \beta & 0
 \end{bmatrix}
\cdot
\begin{bmatrix}
      x-1000 \\
      y-80
 \end{bmatrix}
$$

Em anexo, está a resolução da matriz considerando $a=0,08$, $b=-0,02$, $\alpha = 0,001$ e $\beta = 0,00002$, bem como os valores de $x$ e $y$ previamente estabelecidos.

Ao aplicar o determinante na matriz $det |A- \lambda I|=0$ é obtido $\lambda = \pm i \sqrt{ab}$. Isso significa que o ponto crítico $P_2$ se caracteriza como um centro estável para o sistema linear.

Quando o ponto crítico é um centro do sistema linear correspondente, a estabilidade desse ponto para um sistema quase linear é indeterminada, podendo ser um centro ou uma espiral [2]. Uma forma de determinar as trajetórias do sistema é dividindo a segunda equação pela primeira, como feito no primeiro tópico e resolvendo por equações separáveis.

#### Implementação de algoritmo para análise do modelo de Lotka-Volterra

```python
# IMPORTAÇÕES

import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt
```

```python
a = 1.0
b = 0.1
c = 1.5
d = 0.75
X_f1 = np.array ([c / ( d * b ), a / b])
t = np.linspace (0, 15, 1000)

def dX_dt (X , t = 0):
    return np.array ([a * X [0] - b * X [0] * X [1] , - c * X [1] + d * b * X [0] * X [1]])

# Valores de X0 entre p_1 e p_2
values = np.linspace(0.3, 0.9, 5)
vcolors = plt.cm.rainbow(np.linspace(0.3, 1., len(values))) # definição da paleta de cores

plt.figure(figsize = (15, 10))
plt.rcParams['font.family'] = 'Verdana'

# Representação das trajetórias
for v, col in zip(values, vcolors):
    X0 = v * X_f1 # condição inicial
    X = integrate.odeint( dX_dt, X0, t)
    plt.plot( X[:,0], X[:,1], lw=3.5*v, color=col, label='X0=(%.f, %.f)' % (X0[0], X0[1]))
    
# Definição da malha e do campo de direções
ymax = plt.ylim(ymin=0)[1]
xmax = plt.xlim(xmin=0)[1]
nb_points = 20

x = np.linspace(0, xmax, nb_points) # intervalo de pontos
y = np.linspace(0, ymax, nb_points) # intervalo de pontos

X1 , Y1 = np.meshgrid(x, y) # malha das flexas
DX1, DY1 = dX_dt([X1, Y1]) # taxa de crescimento
M = (np.hypot(DX1, DY1)) # norma da taxa de crescimento
M[ M == 0] = 1 # evita divisão por zero
DX1 /= M # norma da flexa no eixo x
DY1 /= M # norma da flexa no eixo y

# Representação do campo de direções
plt.title('Trajetórias e campo de direções', size=40)
Q = plt.quiver(X1, Y1, DX1, DY1, M, pivot='mid', cmap=plt.cm.jet)
plt.xlabel('Número de presas', size=30)
plt.ylabel('Número de predadores', size=30)
plt.tick_params(labelsize=20)
plt.rcParams['legend.fontsize'] = 20
plt.legend()
plt.grid()
plt.xlim(0, xmax)
plt.ylim(0, ymax)
plt.show()
#plt.savefig('lotka_volterra.png')
```
