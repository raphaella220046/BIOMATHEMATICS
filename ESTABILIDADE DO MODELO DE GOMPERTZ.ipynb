##### A equação diferencial que descreve o Modelo de Gompertz é:

$$\frac{dN}{dt}=-\lambda N ln \frac{N}{K}$$

Os pontos de equilíbrio apresentam um comportamento constante e ocorrem quando a derivada é nula, ou seja, $\frac{dN}{dt}=0$. Para determinar os pontos de equilíbrio pela função, faz-se $f(N)=0$, logo

$$-\lambda N lN \left( \frac{N}{K} \right)=0$$

As duas soluções de equilíbrio para essa situação são $N=0$ e $N=K$. A fim de identificar em que ponto há um crescimento ou decrescimento da função é possível analisar a segunda derivada da equação, ou seja, $\large \frac{d^2N}{dt^2}=\frac{d}{dt} \left( \frac{dN}{dt} \right)= \frac{d}{dt} f(N)$, então

$$f'(N) = -\lambda ln \left( \frac{N}{K} \right) - \lambda N \left( \frac{K}{N} \frac{1}{K} \right)$$

$$f'(N) = -\lambda ln \left( \frac{N}{K} \right) -\lambda \implies -\lambda \left( ln \left( \frac{N}{K} \right) +1 \right)$$

$f'(K)=-\lambda < 0$, logo o ponto de equilíbrio $N=K$ é assintoticamente estável e para $N=0$ é instável. Essa dinâmica pode ser observada no gráfico a seguir.

```python
# Importações
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Parâmetros do modelo de Gompertz
K = 10 # Capacidade de suporte
lamb = 0.1 # Taxa de crescimento

# Função de Gompertz
def modelo_gompertz(t, N):
    dNdt = -lamb * (np.log(N/K)+1)
    return dNdt

# Valores de N (população) e t (tempo)
N_values = np.linspace(0.01, K, 10)
t_values = np.linspace(0, K, 10)

# Definição do campo vetorial
X, T = np.meshgrid(N_values, t_values)
dNdt_values = modelo_gompertz(T, X)

# Representação do campo vetorial
plt.figure(figsize = (15, 10))
plt.quiver(T, X, np.ones_like(dNdt_values), dNdt_values, angles='xy', scale_units='xy', scale=1)
plt.xlabel('Tempo')
plt.ylabel('População (N)')
plt.title('Campo Vetorial do Modelo de Gompertz')
plt.ylim(0, K)
plt.xlim(0, np.max(t_values))
plt.grid()
plt.show()
```

O modelo de Gompertz é uma equação que descreve o crescimento de uma população ao longo do tempo, levando em consideração tanto o crescimento exponencial inicial como a limitação de recursos. O campo vetorial do modelo de Gompertz apresenta duas regiões de importante análise, quando a população está em $N=0$, o $ln$ se torna indefinido, mas ao redor dessa região verifica-se que há um ponto repulsor, evidenciando uma solução instável, enquanto que para $N=k$ o ponto é atrator ou assintoticamente estável. O ponto de estabilidade refere-se ao momento em que a taxa de crescimento da população se torna constante, ou seja, quando o crescimento desacelera e atinge um valor estável de acordo com a capacidade de suporte do meio, se aproximando da solução de equilíbrio em relação ao tempo.
